```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = input("enter the data set:")
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
def perform_statistical_analysis(df):
    """Perform basic statistical analysis."""
    # Calculate mean, median, mode, standard deviation, correlation coefficient
    mean = df.mean()
    median = df.median()
    mode = df.mode().iloc[0]  # Mode might have multiple values
    std_dev = df.std()
    correlation_coefficient = df.corr()

    return mean, median, mode, std_dev, correlation_coefficient
```

```python
def main():
    # Perform statistical analysis
    mean, median, mode, std_dev, correlation_coefficient = perform_statistical_analysis(df)

    # Print results
    print("\nMean:\n", mean)
    print("\nMedian:\n", median)
    print("\nMode:\n", mode)
    print("\nStandard Deviation:\n", std_dev)
    print("\nCorrelation Coefficient:\n", correlation_coefficient)

if __name__ == "__main__":
    main()
```

# **Generating Plots**

# Histogram

```python
df.hist(figsize=(10,10))
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AveragePrice', y='Total Volume', data=df, hue='type', palette='viridis')
plt.title('Scatter Plot of Average Price vs Total Volume')
plt.xlabel('Average Price')
plt.ylabel('Total Volume')
plt.show()
```

# Pie Chart

```python
plt.figure(figsize=(6, 6))
labels = df['type'].value_counts().index
sizes = df['type'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'), wedgeprops=dict(width=0.4))
plt.title('Pie Chart')
plt.show()
```

# Line Chart

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='AveragePrice', data=df[df['type'] == 'organic'], label='Organic', color='green')
sns.lineplot(x='Date', y='AveragePrice', data=df[df['type'] == 'conventional'], label='Conventional', color='purple')
plt.title('Average Price of Avocados Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()
plt.show()
```

# Swarm Plots

```python
plt.figure(figsize=(12, 6))
sns.stripplot(x='region', y='AveragePrice', data= df, palette='viridis', size=4)
plt.title('Strip Plot of Average Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

# Box Plots

```python

plt.figure(figsize=(8, 6))
sns.boxplot(y='AveragePrice', data= df, palette='viridis')
plt.title('Box Plot of Average Price')
plt.xlabel('Average Price')
plt.xticks(rotation=90)
plt.show()



plt.figure(figsize=(8, 6))
sns.boxplot(y='Total Volume', data= df, palette='viridis')
plt.title('Box Plot of Total Volume')
plt.xlabel('Total Volume')
plt.xticks(rotation=90)
plt.show()

```

# Steam Leaf Plot

```python
plt.figure(figsize=(13, 6))
plt.stem(df['AveragePrice'], linefmt='-', markerfmt='o', basefmt=' ', use_line_collection=True)
plt.title('Stem and Leaf Plot of Average Price')
plt.xlabel('Index')
plt.ylabel('Average Price')
plt.show()
```



#Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['AveragePrice'], palette='viridis')
plt.title('Violin Plot of Average Price')
plt.xlabel('Average Price')
plt.show()
```

# Doughnut Chart

```python
doughnut_data = df['type'].value_counts()

# Create a doughnut chart
plt.figure(figsize=(8, 8))
plt.pie(doughnut_data, labels=doughnut_data.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'], wedgeprops=dict(width=0.4))
plt.gca().add_artist(plt.Circle((0, 0), 0.6, fc='white'))

plt.title('Doughnut Chart of Avocado Types')
plt.show()
```
